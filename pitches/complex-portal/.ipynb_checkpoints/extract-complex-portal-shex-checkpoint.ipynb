```python
# from wikidataintegrator import wdi_core
subsets = [] 
sets =[]
i=0
for gbif in gbifIds:
  sets.append(gbif)
  if i % 500 == 0:
    subsets.append(sets)
    sets = []
  i+=1

results = []
for subset in subsets:
  query = "SELECT ?taxon ?taxonLabel ?inat WHERE {VALUES ?gbifid {"
  query += " ".join(subset)+ "}"
  query += " ?taxon wdt:P685 ?ncbi_taxonID ; wdt:P846 ?gbifid ; wdt:P3151 ?inat . SERVICE wikibase:label { bd:serviceParam wikibase:language \"[AUTO_LANGUAGE],en\". }}"
  print(query)
  results.append(wdi_core.WDFunctionsEngine.execute_sparql_query(query, as_dataframe=True))
  



```

```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
concatdf=pd.concat(results)
concatdf
```

```python
from rdflib import Graph

typespecimenGraph = Graph()

for index, row in concatdf.iterrows():
  print(row["taxon"])
  typespecimenGraph.parse(row["taxon"], format="turtle")
```

```python
qres = typespecimenGraph.query("SELECT DISTINCT ?type WHERE {?taxon wdt:P31 ?type}")
target_classes = []
for row in qres: target_classes.append(str(row["type"]))
```

```python
target_classes
```

```python
from shexer.shaper import Shaper

shex_target_file = "wikidata_type_specimens.shex"

namespaces_dict = {
            "http://www.w3.org/2000/01/rdf-schema#": "rdfs",
            "http://www.wikidata.org/prop/": "p",
            "http://www.wikidata.org/prop/direct/": "wdt",
            "http://www.wikidata.org/entity/": "wd",
            "http://www.w3.org/2001/XMLSchema#": "xsd",
            "http://www.w3.org/1999/02/22-rdf-syntax-ns#": "rdf",
            "http://www.w3.org/XML/1998/namespace": "xml",
            "http://wikiba.se/ontology#": "wikibase",
            "http://schema.org/": "schema",
            "http://www.w3.org/2004/02/skos/core#": "skos",
            "http://schema.org/": "schema",
            "http://www.w3.org/2004/02/skos/core#": "skos",
            "http://wikiba.se/ontology#": "wikibase" ,
        }

namespaces_to_ignore = [
    "http://www.wikidata.org/prop/",                    
]


typespecimenGraph.serialize(destination="input_graph.ttl", format="turtle")
input_nt_file = "input_graph.ttl"

shaper = Shaper(target_classes=target_classes,
                input_format='turtle',
                depth_for_building_subgraph=1,
                graph_file_input=input_nt_file,
                namespaces_dict=namespaces_dict,  # Defaults: no prefixes
                namespaces_to_ignore=namespaces_to_ignore,
                instantiation_property="http://www.wikidata.org/prop/direct/P31") 


shaper.shex_graph(output_file=shex_target_file)

print("Done!")

```
