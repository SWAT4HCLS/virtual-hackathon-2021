```python
from wikidataintegrator import wdi_core
query = 'SELECT ?item ?itemLabel ?cpid WHERE {?item wdt:P7718 ?cpid . SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}'
concatdf = wdi_core.WDFunctionsEngine.execute_sparql_query(query, as_dataframe=True)
concatdf
```

```python
from rdflib import Graph

complexPortalGraph = Graph()

for index, row in concatdf.iterrows():
  print(row["item"])
  complexPortalGraph.parse(row["item"], format="turtle")
```

```python
qres = complexPortalGraph.query("SELECT DISTINCT ?type WHERE {?complex wdt:P31 ?type}")
target_classes = []
for row in qres: target_classes.append(str(row["type"]))
```

```python
target_classes
```

```python
from shexer.shaper import Shaper

shex_target_file = "wikidata_complex_portal.shex"

namespaces_dict = {
            "http://www.w3.org/2000/01/rdf-schema#": "rdfs",
            "http://www.wikidata.org/prop/": "p",
            "http://www.wikidata.org/prop/direct/": "wdt",
            "http://www.wikidata.org/entity/": "wd",
            "http://www.w3.org/2001/XMLSchema#": "xsd",
            "http://www.w3.org/1999/02/22-rdf-syntax-ns#": "rdf",
            "http://www.w3.org/XML/1998/namespace": "xml",
            "http://wikiba.se/ontology#": "wikibase",
            "http://schema.org/": "schema",
            "http://www.w3.org/2004/02/skos/core#": "skos",
            "http://schema.org/": "schema",
            "http://www.w3.org/2004/02/skos/core#": "skos",
            "http://wikiba.se/ontology#": "wikibase" ,
        }

namespaces_to_ignore = [
    "http://www.wikidata.org/prop/",                    
]


complexPortalGraph.serialize(destination="input_graph.ttl", format="turtle")
input_nt_file = "input_graph.ttl"

shaper = Shaper(target_classes=target_classes,
                input_format='turtle',
                depth_for_building_subgraph=1,
                graph_file_input=input_nt_file,
                namespaces_dict=namespaces_dict,  # Defaults: no prefixes
                namespaces_to_ignore=namespaces_to_ignore,
                instantiation_property="http://www.wikidata.org/prop/direct/P31") 


shaper.shex_graph(output_file=shex_target_file)

print("Done!")

```
